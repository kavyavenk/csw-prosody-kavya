```python
import pandas as pd
```

```python
with_strat = pd.read_csv('MaSac transcripts with strategy annotations.csv')
with_strat.head()
```

```python
with_paths = pd.read_csv('masac_data_compiled.csv')
with_paths.head()
```

```python
with_paths = with_paths[['name', 'Transcript', 'English Translation', 'LID_tags', 'Language']]
with_paths.head()
```

```python
len(with_paths) == len(with_strat)
```

```python
len(with_paths)
```

```python
len(with_strat)
```

```python
with_path_transcript = list(with_paths['Transcript'])
with_strat_transcript = list(with_strat['Transcript'])
strat_annotations = list(with_strat['Type of code-switching'])
```

```python
path_linked_annotations = []
```

```python
for transcript_p in with_path_transcript:
    found = False
    for i in range(len(with_strat_transcript)):
        transcript_a = with_strat_transcript[i]
        if transcript_p == transcript_a:
            path_linked_annotations.append(strat_annotations[i])
            found = True
            break
    if found == False:
        path_linked_annotations.append('TODO')
```

```python
len(path_linked_annotations)
```

```python
path_linked_annotations.count('TODO')
```

```python
from collections import defaultdict
```

```python
annotations_dict = defaultdict()
```

```python
for label in path_linked_annotations:
    if label not in annotations_dict.keys():
        annotations_dict[label] = 1
    else:
        annotations_dict[label] += 1

annotations_dict
```

```python
with_paths['Type of code-switching'] = path_linked_annotations
with_paths.head()
```

```python
with_paths.to_csv('masac_data_compiled_with_strategies.csv')
```
